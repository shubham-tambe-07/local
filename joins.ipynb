```python
empdata = [(101,'keshav',3242),(102,'amit',42),(104,'ram',42),(105,'shyam',42)]
deptdata = [(101,'hr'),(102,'it'),(501,'sales'),(104,'hr'),(500,'it')]
```

```python
df1  = spark.createDataFrame(empdata,['id','name','salary'])
df2 = spark.createDataFrame(deptdata,['id','dept'])
```

```python
df1.join(df2,['id']).show()
```

```python
df1.join(df2,on="id",how="inner").show()
```

```python
from pyspark.sql.functions import broadcast
```

```python
df1.join(broadcast(df2),on="id",how="inner").explain()
```

```python
df1.join(df2.hint("cartesian")).show()
```

```python
df1.join(df2.hint("broadcast"),df1.id<df2.id,how="inner").show()
```

```python
df1.show()
df2.show()
```
